```python
import os

train_txt_path = "/nvme/data/video_45/train.txt"
test_txt_path = "/nvme/data/video_45/test.txt"

train = "/nvme/data/video_45/train"
test = "//nvme/data/video_45/test"

def creat_set_txt(path, data_path):
    file = open(path, 'w')
    videos = os.listdir(data_path)
    for video in videos:
        file.write(video+'\n')
    file.close()

creat_set_txt(test_txt_path, test)
creat_set_txt(train_txt_path, train)
```

```python
import SimpleITK as sitk
import numpy as np
import os
path = r'D:\TransUNet\data\oct\train_npz'
path0 =r'D:\first_set\tr_label'
labs = os.listdir(path)
for lab in labs:
    label_path = os.path.join(path, lab)
    data = np.load(label_path)
    label = data['label']
    if label.sum() == 0:
        name = lab.replace('.npz','.nii.gz')
        source_path = os.path.join(path0, name)
        if os.path.exists(source_path):
            nii = sitk.ReadImage(source_path)
            lab_np = sitk.GetArrayFromImage(nii)
            if lab_np.sum()==0:
                print(source_path)
```
