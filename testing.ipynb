```python
import cv2 as cv
import numpy as np
image = cv.imread(r'D:\oct\1.jpg')
dst	=	cv.warpPolar(image, (159,999),(159,159), 159, cv.INTER_LINEAR + cv.WARP_POLAR_LINEAR)
cv.imwrite(r"D:\oct\lined.jpg", dst)


lined_inversed = cv.warpPolar(dst, (320,320), (159,159), 159, cv.WARP_POLAR_LINEAR + cv.WARP_INVERSE_MAP)
cv.imwrite(r"D:\oct\lined_inversed.jpg", lined_inversed)


mask = np.zeros_like(dst)
mask[900:930,:,:] = 255
mask_shadow = cv.GaussianBlur(mask,(3, 99),0)
masked_lined = cv.subtract(dst,mask_shadow)
cv.imwrite(r"D:\oct\masked_lined.jpg", masked_lined)


shadow_inversed = cv.warpPolar(masked_lined, (320,320), (159,159), 159, cv.WARP_POLAR_LINEAR + cv.WARP_INVERSE_MAP)
cv.imwrite(r"D:\oct\shadow_inversed.jpg", shadow_inversed)


circle_mask = np.zeros_like(shadow_inversed)
c_mask = cv.circle(circle_mask,(155,155), 159, (1,1,1), -1)
target = shadow_inversed * c_mask

cv.imwrite(r"D:\oct\masked_inversed.jpg", target)
cv.waitKey()
cv.destroyAllWindows()
```

```python
def cart2poler(image, flag=cv.INTER_LINEAR):
    h, w = image.shape[:2]
    circle_radius = int(h / 2)
    polar_shape = int(circle_radius / 1), int(2 * circle_radius * 3.14)
    center = (circle_radius, circle_radius)
    polar_image = cv.warpPolar(image, polar_shape, center, circle_radius, flag + cv.WARP_POLAR_LINEAR)
    name = r"D:\oct\\"+ str(flag) + r'.jpg'
    cv.imwrite(name ,polar_image)
    return polar_image
image = cv.imread(r'D:\oct\0.png')
cart2poler(image,flag=cv.INTER_BITS2)
```
